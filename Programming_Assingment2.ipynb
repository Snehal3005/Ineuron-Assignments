1.	Write a Python program to convert kilometers to miles?

```python
# input from the user
kilometers = float(input("Enter value in kilometers: "))

# conversion
con = 0.621371

# calculate miles
miles = kilometers * con

print(f'{kilometers:{5}.{4}} km is equal to {miles:{4}.{4}} miles')
```

2.	Write a Python program to convert Celsius to Fahrenheit?

```python
# input from the user
celsius = float(input("Enter value in celsius: "))

# conversion
fahrenheit = 9.0/5.0 * celsius + 32

print(f'{celsius:{5}.{4}} celsius is equal to {fahrenheit:{4}.{4}} F')
```

3.	Write a Python program to display calendar?

```python
import calendar

year = int(input("Enter year: "))
month = int(input("Enter month: "))

print(calendar.month(year, month))
```

4.	Write a Python program to solve quadratic equation?

```python
# ax2 + bx + c = 0 this is standard form of quadratic equation
# accept a,b,c from user

a = int(input("Enter value of a: "))
b = int(input("Enter value of b: "))
c = int(input("Enter value of c: "))

#find roots using formula
x1 = (-b + (b**2 - (4*a*c))**0.5)/ (2*a)
x2 = (-b - (b**2 - (4*a*c))**0.5)/ (2*a)

print(f'Roots are: {x1} and {x2}')
```

5. Write a Python program to swap two variables without temp variable?

```python
# accept input from user
var1 = input('Enter first value::')
var2 = input('Enter second value::')

#assignment at the same time

var1,var2 = var2,var1

#print variable values after swapping

print(f'value of first variable after swapping is {var1}')
print(f'value of second variable after swapping is {var2}')
```
