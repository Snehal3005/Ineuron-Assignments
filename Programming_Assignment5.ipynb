1. Write a Python Program to Find LCM?

```python
# number of elements
import math    

def lcm_n(numbers):
    res = numbers[0]
    for i in range(1, len(numbers)):
        gcd = math.gcd(res,numbers[i])
        res = res * numbers[i]
        res = int(res/gcd)
    
    return res

n = int(input('Enter number of elements:'))
numbers = []

for i in range(0,n):
    n1 = int(input('Enter number:'))
    numbers.append(n1)
if n > 0:
    lcm = lcm_n(numbers)
    print(f'LCM of {numbers} is {lcm}')    
else:
    print('Please enter valid number ')
```

2. Write a Python Program to Find HCF?

```python
import math

n = int(input('Enter number of elements:'))
numbers = []

for i in range(0,n):
    n1 = int(input('Enter number:'))
    numbers.append(n1)
    
print(f'HCF of {numbers} is {gcd_n(numbers)}')

def gcd_n(numbers):
    res = numbers[0]
    for i in range(1,len(numbers)):
        res = math.gcd(numbers[i],res)
    return res
```

3. Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

```python
num = int(input('Enter a decimal number::'))
output = ''

def convert(num,base):
    remainder = ''    
    if base == 2 and (num == 0 or num == 1):
        remainder = num
    elif base == 8 and (num >= 0 and num <= 7):    
        remainder = num
    elif base == 16 and num >= 0 and num <= 9:
        remainder = num        
    elif base == 16 and num == 10:
        remainder = 'A'
    elif base == 16 and num == 11:
        remainder = 'B'
    elif base == 16 and num == 12:
        remainder = 'C' 
    elif base == 16 and num == 13:
        remainder = 'D' 
    elif base == 16 and num == 14:
        remainder = 'E' 
    elif base == 16 and num == 15:
        remainder = 'F'
    else:        
        while num != 0:
            mod = int(num % base)
            num = int(num / base)
            if mod == 10:
                remainder += 'A'
            elif mod == 11:
                remainder += 'B'
            elif mod == 12:
                remainder += 'c'
            elif mod == 13:
                remainder += 'D'
            elif mod == 14:
                remainder += 'E'
            elif mod == 15:
                remainder += 'F'
            else:
                remainder += str(mod)

        remainder = remainder[::-1]
    return remainder

output = convert(num,2)
print(f'Binary equivalent of {num} is {output}')   

output = convert(num,8)
print(f'octal equivalent of {num} is {output}')   

output = convert(num,16)
print(f'Hexadecimal equivalent of {num} is {output}')   

```

5. Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

```python
n1 = int(input('Enter number 1:'))
n2 = int(input('Enter number 2:'))

if n1 >= 0 and n2 >= 0:
    print('\n1.Addition \n2.Multiplication \n3.Subtraction \n4.Division \n')
    operation = input('Enter what operation would you like to perform(1/2/3/4):')
    if operation == '1':
        print(f'Addition of {n1} and {n2} is {n1+n2}')
    elif operation == '2':
        print(f'Multiplication of {n1} and {n2} is {n1*n2}')    
    elif operation == '3':
        print(f'Subtraction of {n1} and {n2} is {n1-n2}')
    elif operation == '4':
        if n2 != 0:
            print(f'Division of {n1} and {n2} is {n1/n2}')
        else:
            print('Can not divide by 0')
    else:
        print('Please enter valid option')
```
