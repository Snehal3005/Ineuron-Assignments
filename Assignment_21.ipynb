1. Add the current date to the text file today.txt as a string.

```python
import datetime

file = open("today.txt","w+")
today = str(datetime.date.today())
file.write(today)
file.close()
```

2. Read the text file today.txt into the string today_string

```python
file = open("today.txt","r")
file.seek(0)
today_string = file.read()
print(today_string)
```

3. Parse the date from today_string.

```python
date = datetime.datetime.strptime(today_string,"%Y-%m-%d")
print(date)
```

6. Use multiprocessing to create three separate processes. Make each one wait a random number of seconds between one and five, print the current time, and then exit.

# following code if saved as .py file is working fine but not giving expected result in jupyter notebook. 


```python
from multiprocessing import Process
import time
import random
import datetime

def is_even(a):
    print('\n starting even_odd function')
    wait_time = random.choice(range(1,5))
    time.sleep(wait_time)
    print(datetime.datetime.now())
    if a % 2 == 0:
        print("number is Even")
    else:
        print("number is Odd")
    print('finishing even_odd function')


def square(n):
    print('starting sqaure function')
    wait_time = random.choice(range(1,5))
    time.sleep(wait_time)
    print(datetime.datetime.now())
    print("Square: {}".format(n * n))
    print('finishing square fun')


def cube(n):
    print('Starting cube function')
    wait_time = random.choice(range(1,5))
    time.sleep(wait_time)
    print(datetime.datetime.now())
    print("Cube: {}".format(n**3))
    print('finishing Cube fun')

if __name__ == '__main__':

    pro1 = Process(target=is_even,args=(2,))
    pro2 = Process(target=square,args=(3,))
    pro3 = Process(target=cube,args=(3,))

    pro1.start()
    pro2.start()
    pro3.start()

    pro1.join()
    pro2.join()
    pro3.join()

    print("Processes Completed!")
```
