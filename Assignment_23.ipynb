1. What is the result of the code, and why?

    def func(a, b=6, c=8):

    print(a, b, c)

    func(1, 2)

```python
def func(a, b=6, c=8):
    print(a, b, c)

func(1, 2)

# output is 1,2,8 because the value of c is default value.Default values means that the function
# argument will take that value if no argument value is passed during the function call
```

4. What does this code print, and why?

    def func(a, **kargs):

        print(a, kargs)

    func(a=1, c=3, b=2)


```python
def func(a, **kargs):
    
    print(a, kargs)

func(a=1, c=3, b=2)


# A keyword argument is where you provide a name to the variable as you pass it into the function.
# It is like dictionary that maps each keyword to the value that we pass alongside it. 
# Here b and c are keys with values 2 and 3 respectively
```

5. What gets printed by this, and explain?

    def func(a, b, c=8, d=5): 

        print(a, b, c, d)

    func(1, *(5, 6))


```python
def func(a, b, c=8, d=5):
    print(a, b, c, d)

func(1, *(5, 6))

# Here a = 1 which is the value passed in a function call
# variable b and c are *args which are always passed in tuple form abd default value of c is replaced with passed value
# variable d has default value 5
```

6. what is the result of this, and explain?

def func(a, b, c): 

    a = 2; b[0] = 'x'; c['a'] = 'y'

    l=1; m=[1]; n={'a':0}

func(l, m, n)


```python
def func(a, b, c):
    a = 2
    b[0] = 'x'
    c['a'] = 'y'
    
l=1
m=[1]
n={'a':0}

func(l, m, n)
print(l,m,n)


# here, scope of varible a is local to function func() and not affecting original value l. Once we leave the scope 
# of function func() , a=”new value” is no longer in the name space, and the value that refers to was never changed. This is 
# call by value because type of a in integer which is immutable

# This is call by reference as list and dictionaries are mutable because when their values are changed inside the function, 
# then it will also be reflected outside the function.

# variable b is list, b[0] is referring to same memory location and thats why value of m is also modified
# c is dictionary, initially the key 'a' holds value 0 but is reassigned with new value because memory location is same
```
