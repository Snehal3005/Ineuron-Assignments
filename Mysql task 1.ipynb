```python
# Write Carbon nanotype csv file to mysql 
# data https://archive.ics.uci.edu/ml/datasets/Carbon+Nanotubes

import mysql.connector as connection
import csv
try:
   
    mydb = connection.connect(host="127.0.0.1",user="root", passwd="mysql",use_pure=True, database = 'snehal')
    cursor = mydb.cursor()

    with open('C://Users/panka/Downloads/carbon_nanotubes.csv', 'r') as csvfile:
        csvreader = csv.reader(csvfile,delimiter=';')

        # Read from line 2, skip column names
        header = next(csvreader)

        # Create table
        cursor.execute('create table IF NOT EXISTS Carbon4 (c1 varchar(10),c2 varchar(10),c3 varchar(10),c4 varchar(10),c5 varchar(10),c6 varchar(10),c7 varchar(10),c8 varchar(10))')

        for row in csvreader:
            row = str(row).replace('[','').replace(']','') # to make it in the correct form for insert
            cursor.execute(f'INSERT INTO Carbon4 values ({row})')
            mydb.commit()
        print("Values inserted sucessfully!!")
        cursor.close()
        mydb.close()
except Exception as e:
    print(str(e))

```
