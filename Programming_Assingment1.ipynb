2. Write a Python program to do arithmetical operations addition and division.?

```python
# accept input from user
number1 = int(input('Enter first number::'))
number2 = int(input('Enter second number::'))

#perform addition and print result
print(f'Result after adding {number1} and {number2} is {number1+number2}')

# check if denominator is not 0 then perform division
if(number2 != 0):
    print(f'Result after dividing {number1} by {number2} is {number1/number2}')
else:
    print('please enter a non zero value for denominator')
```

3.	Write a Python program to find the area of a triangle?

```python
# accept base and height of triangle from user
base = int(input('Enter base of a triangle::'))
height = int(input('Enter height of a triangle::'))

#calcultae area of triangle(hb/2)
area = (1/2) * base * height

# print area
print(f'Area of a given triangle is {area}')
```

4.	Write a Python program to swap two variables?

```python
# accept input from user
var1 = input('Enter first value::')
var2 = input('Enter second value::')

#create a temporary variable which will hold old value of variable 1
temp = var1
var1 = var2
var2 = temp

#print variable values after swapping

print(f'value of first variable after swapping is {var1}')
print(f'value of second variable after swapping is {var2}')
```
