4. What output does this code produce? Why?

X = "iNeuron"

def func():

    global X
    
    X = "NI"

func()

print(X)

Global keyword allows a user to modify a variable outside of the current scope. So here, variable X is modified with value("NI") by using global keyword thus, print(X) function is printing modified value.

```python
X = "iNeuron"

def func():
    global X
    X = "NI"

func()
print(X)
```

    The nonlocal keyword is used to work with variables inside nested functions, where the variable should not belong to the inner function. Use the keyword nonlocal to declare that the variable is not local.

```python
def func(): 
    X = 'NI'

    def nested():
        nonlocal X 
        X = 'Spam'

    nested()
    print(X)

func()
```
